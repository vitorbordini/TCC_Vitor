```

def brightness(X):
    """
    Changing the brighness of a image using power-law gamma transformation.
    Gain and gamma are chosen randomly for each image channel.
    
    Gain chosen between [0.8 - 1.2]
    Gamma chosen between [0.8 - 1.2]
    
    new_im = gain * im^gamma
    """
    
    X_new = np.zeros(X.shape)
    for c in range(X.shape[-1]):
        im = X[:,:,:,c]        
        gain, gamma = (1.2 - 0.8) * np.random.random_sample(2,) + 0.8
        im_new = np.sign(im)*gain*(np.abs(im)**gamma)
        X_new[:,:,:,c] = im_new 
    
    return X_new
```

```
import torchio as tio
import numpy as np
import matplotlib.pyplot as plt

layer = 50
img = np.expand_dims(nib.load(list_IDsT1[0]).get_fdata(),axis=0).astype(np.uint8)
elastic = tio.RandomElasticDeformation(num_control_points=(7, 7, 7),locked_borders=2)
elastimg = elastic(img)

scale  = tio.RandomAffine(scales=(0.9,1.2),degrees=0, image_interpolation='nearest')
scaled = scale(img)

rot = tio.RandomAffine(scales=(0.002),degrees=(0,35), image_interpolation='nearest')
rotation = rot(img)

bright = brightness(img)

flip = tio.RandomFlip(axes=1,flip_probability=1.)
flipped = flip(img)
fontsize = 30
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(20, 20))
#ax[0][0].imshow(np.squeeze(img[:,:,:,layer]))
#ax[0][0].set_title("Original", fontsize=fontsize)

ax[0][0].imshow(np.squeeze(elastimg[:,:,:,layer]))
ax[0][0].set_title("Deformação elástica", fontsize=fontsize)

ax[0][1].imshow(np.squeeze(scaled[:,:,:,layer]))
ax[0][1].set_title("Escala", fontsize=fontsize)

ax[0][2].imshow(np.squeeze(bright[:,:,:,layer]))
ax[0][2].set_title("Ajuste de brilho", fontsize=fontsize)

ax[1][0].imshow(np.squeeze(img[:,20:120,20:120,layer]))
ax[1][0].set_title("Recorte", fontsize=fontsize)

ax[1][1].imshow(np.squeeze(flipped[:,:,:,layer]))
ax[1][1].set_title("Espelhamento", fontsize=fontsize)

ax[1][2].imshow(np.squeeze(rotation[:,:,:,layer]))
ax[1][2].set_title("Rotação", fontsize=fontsize)
for i in range(len(ax)):
  for j in range(len(ax[0])):
    ax[i][j].axis('off')

plt.savefig("Augmentation.eps")
```
